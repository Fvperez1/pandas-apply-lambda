# CHALLENGE1


```python
def filtro(x):
    if x < 1000:
        y = 1
    elif x >= 1000 and  x < 10000:
        y = 2
    elif x >= 10000 and x < 99999:
        y = 3
    elif x >= 100000 and x < 999999:
        y = 4
    else:
        y = 5
    return y
```

```python
imdb['bin'] = imdb.apply(lambda x: filtro(x['Votes']), axis=1)
imdb.head()
```

# Challenge 2

```python
def media(x,y):
    return x/y
```

```python
imdb['Rev/min'] = imdb.apply(lambda x: media(x['Revenue (Millions)'], x['Runtime (Minutes)']), axis=1)
imdb.head()
```

# Challenge3

```python
def nuevoRanking(x,y):
    if 'Comedy' in y:
        x -= 1
    if 'Thriller' in y:
        x += 1
    return x
    
```

```python
imdb['New Ranking'] = imdb.apply(lambda x: nuevoRanking(x['Rating'], x['Genre']), axis=1)
imdb.head()
```

# Challenge4

```python
def challenge4(x,y):
    z = 0
    for i in x:
        z += ord(i)
    z /= y
    z = int(z)
    primo = "Es primo"
    for n in range(2,z):
        if z % n ==0:
            primo = "No es primo"
  
    return primo
```

```python
imdb['Is Primo'] = imdb.apply(lambda x: challenge4(x['Title'], x['Votes']), axis=1)
imdb.head()
```
